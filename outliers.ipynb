```python
import pandas as pd
import numpy as np
states_all_raw = pd.read_csv('data/states_all.csv')
```

```python
df = states_all_raw.groupby('STATE').apply(
    lambda grp: grp.interpolate(method='linear', 
                                limit_direction='both', 
                                axis=0)).dropna().reset_index(drop=True)
display(df)
```

# 1

```python
from scipy.stats import zscore

df
```

```python
def outliercheck(df, thresh):
    z_rev = zscore(df['TOTAL_REVENUE'])
    z_exp = zscore(df['TOTAL_EXPENDITURE'])
    rev_outliers = np.where((z_rev < -thresh) | (z_rev > thresh))[0]
    exp_outliers = np.where((z_exp < -thresh) | (z_exp > thresh))[0]
    outliers = np.unique(np.append(rev_outliers,exp_outliers))
    return outliers, rev_outliers.size, exp_outliers.size, outliers.size
for i in range(1,7):
    print('threshold=',i,outliercheck(df, i)[1:])
```

There are outlier values for threshold values of |z| = 1, 2, 3, 4, and 5
